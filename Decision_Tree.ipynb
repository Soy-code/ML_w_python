```
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn import metrics
```

```
# importing datset
iris = sns.load_dataset("iris")
X = iris.drop("species", axis = 1)
y = iris['species']
```

```
# factorization 
classle = LabelEncoder()
y = classle.fit_transform(iris['species'].values) 
np.unique(y)  # 0, 1, 2로 factorized 됐음.
```

```
# split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
                                                    random_state = 1, stratify = y)
```

```
# standardization
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```
# Classification Tree model fitting
# with depth = 3
decision_tree = tree.DecisionTreeClassifier(criterion = "gini", max_depth = 3, random_state = 1)
# with depth = 4
decision_tree2 = tree.DecisionTreeClassifier(criterion = "gini", max_depth = 4, random_state = 1)
# with depth = 5
decision_tree3 = tree.DecisionTreeClassifier(criterion = "gini", max_depth = 5, random_state = 1)

# fitting with training data 
decision_tree.fit(X_train, y_train)
y_train_pred = decision_tree.predict(X_train)
y_test_pred = decision_tree.predict(X_test)

decision_tree2.fit(X_train, y_train)
y_train_pred2 = decision_tree2.predict(X_train)
y_test_pred2 = decision_tree2.predict(X_test)

decision_tree3.fit(X_train, y_train)
y_train_pred3 = decision_tree3.predict(X_train)
y_test_pred3 = decision_tree3.predict(X_test)
```

```
# tree plot 1
tree.plot_tree(decision_tree.fit(X_train, y_train))
```

```
# tree plot 2
tree.plot_tree(decision_tree2.fit(X_train, y_train))
```

```
# tree plot 3
tree.plot_tree(decision_tree3.fit(X_train, y_train))
```

```
# accuracy 
print(metrics.accuracy_score(y_train, y_train_pred))
print(metrics.accuracy_score(y_test, y_test_pred))
print("==========================================")
print(metrics.accuracy_score(y_train, y_train_pred2))
print(metrics.accuracy_score(y_test, y_test_pred2))
print("=========================================")
print(metrics.accuracy_score(y_train, y_train_pred3))
print(metrics.accuracy_score(y_test, y_test_pred3))
```
