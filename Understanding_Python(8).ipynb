```
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
len(x)
plt.plot(x, np.sin(x))
plt.show()  # 그림의 끝에 입력 
            # 근데 이거 안 써도 그림 출력은 되는 것 같은디
```

```
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.show()
```

```
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])  # 
plt.show()
```

```
# 그래프 스타일 변경 
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html 
# 다음 링크에서 matplot 그래프 스타일에 대한 정보
plt.plot(x, np.sin(x), 'b-o')  # 파란색 동그라미로 표시한 후 줄로 연결
plt.show()
```

```
plt.style.use('seaborn-white')
plt.plot(x, np.sin(x), 'b', label = 'sin') # blue 
plt.plot(x, np.cos(x), label = 'cos')
plt.title('sine - cosine curve')
plt.legend()
plt.show()
```

```
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = "cos")
plt.legend(loc = "upper right")
plt.xlabel('x')
plt.ylabel('sine and cosine curves')
plt.title('sine - cosine curve')
plt.show()
```

```
# 두 개 이상의 그래프를 하나로 묶어서 출력
import matplotlib.pyplot as plt
import numpy as np
n = np.linspace(0, 10, 100)
plt.figure()  # figure  시작
plt.subplot(2, 1, 1)  # subplot(행, 열, 몇 번째 plot인지)
plt.plot(x, np.sin(x), 'b--', label = 'sine') # blue & 점선으로 이어라? 
plt.plot(x, np.cos(x), label = "cosine")
plt.title('sine and cosine curve')
plt.legend(loc = "best")  # 그냥 최적에 자리에 레전드 입혀주나봄
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), label = 'right')
plt.legend(loc = 'best')
plt.show()
```

```
# seaborn을 이용ㅇ한 그래프 그리기
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()   # 위에랑 그림 스타일이 다른 건가 
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'b--', label = 'sine')
plt.legend(loc = 'best')
plt.show()
```

```
# 히스토그램 그리기
import pandas as pd
bn = np.random.multivariate_normal([0, 0], [[5, 1], [1, 2]], size = 1000)  # 2 variables
bn = pd.DataFrame(bn, columns = ['x', 'y'])  
print(bn.head())

for i in 'xy' :
  plt.hist(bn[i], density = True, alpha = 0.6) # density = True : y축 값을 백분율로
                                               # default는 density = False임 
                                               # alpha : 불투명도, 1에 가까울수록 불투명
```

```
# kernel density estimation을 이용하여 그리기
for i in 'xy' :
  sns.kdeplot(bn[i], shade = True)   # kernel density function

sns.distplot(bn['x'])
plt.show()
```

```
sns.distplot(bn['y'])
```

```
sns.kdeplot(bn)  # kde 추정치? 이게 뭐지 ★★★★★
```

```
# kde와 주변분포 그리기
sns.jointplot(bn['x'], bn['y'], kind = 'kde')
```

```
# 회귀선과 주변 분포 그ꈰ기
sns.jointplot(x = 'x', y = 'y', data = bn, kind = 'reg')
# sns.jointplot(bn['x'], bn['y'], data = bn, kind = 'reg') 
```

```
# 각 pair에 대한 산포도 그리기
iris = sns.load_dataset('iris')
iris.head()
```
