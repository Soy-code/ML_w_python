Plotly에서 plotly.express와 plotly.graph_objects를 이용해서 그림을 생성함. <br>
(근데 github에 업로드할 때 plotly의 static한 그래프 이미지가 업로드가 안 됨. 이거 해결해봐야 할 듯)

```
import plotly.express as px
tips_df = px.data.tips()
tips_df.head()
```

```
tips_df.columns
```

```
fig = px.histogram(tips_df, x = "total_bill", y = "tip", 
                   color = "smoker", marginal = "rug", 
                   hover_data = tips_df.columns)  
# hover_data : 데이터의 세부 정보를 추가적으로 보여주는 팝업 정보창 (interactive하게)
fig.update_layout(title_text = "Histogram", title_x = 0.5) 
# title_x = 0.5 : title을 가운데에
fig.show()
```

```
fig = px.box(tips_df, x = "time", y = 'tip', color = "sex")
fig.update_layout(title_text = "Boxplot", 
                  title_x = .5, title_y = .5, # plot 가운데에 title 놓음. 
                  title_font_size = 30)
fig.show()
```

```
fig = px.violin(tips_df, x = 'sex', y = 'tip', color = 'smoker', 
                box = True, points = "all", hover_data = tips_df.columns)
fig.show()
# violin은 distribution을 대칭적으로 하여 보여주는 그래프. 
# 근데 여기서 box = True 옵션을 넣어줘서 boxplot도 추가됨
# points 또한 넣어줌 
```

```
import numpy as np
pop_df = px.data.gapminder()
df_1 = np.unique(pop_df['country'])  # country의 원소의 종류를 보여줌
print(df_1)
```

```
df_korea = pop_df.query("country == 'Korea, Rep.'")
df_korea.head()  # 한국의 경우만을 출력
```

```
fig = px.line(df_korea, x = "year", y = "lifeExp",
              title = "Life Expectancy in Korea")
fig.show()
```

```
df_asia = pop_df.query("continent == 'Asia'")
fig = px.line(df_asia, x = "year", y = "lifeExp", color = "country")  # asia 내에서 국가마다
fig.show()
```

```
import plotly.graph_objects as go 
X = df_korea['year']
Y = df_korea['lifeExp']
fig = go.Figure(data = go.Scatter(x = X, y = Y))
fig.show()
```

```
import numpy as np
import plotly.graph_objects as go
np.random.seed(0)
N = 100 
x1 = np.linspace(0, 10, N)  # 0과 10 사이를 100개로 분할
y1 = 5 + x1 + np.random.randn(N)  # 절편이 5이고 coef가 1인 linear regression.
                                  # 마지막엔 noise를 추가
y2 = x1 + np.random.randn(N) 
y3 = -5 + x1 + np.random.normal(0, 2, N)

fig = go.Figure()
fig.add_trace(go.Scatter(x = x1, y = y1, mode = "lines", name = "line"))
fig.add_trace(go.Scatter(x = x1, y = y2, mode = "lines + markers", name = "line + marker"))
fig.add_trace(go.Scatter(x = x1, y = y3, mode = "markers", name = "marker"))
fig.show()
```

```
import numpy as np
import plotly.graph_objects as go
import plotly.offline as pyo

np.random.seed(0)
N = 100

x1, x2 = np.meshgrid(np.linspace(0, 2, N), np.linspace(0, 2, N))  # 각각 N X N의 정방행렬을 만듦
print(x1)
print('--------------------------------------------------------------------------------------')
print(x2)
# 근데 행/열 방향이 다르네
```

```
y1 = 5 + x1 + 2*x2  # y1 또한 N X N matrix 
y2 = y1.copy() + np.random.normal(0, 2, (N,N))  # y1에다가 추가적으로 noise를 줌
```

```
# 얘는 surface를 만듦
surface_1 = go.Surface(x = x1, y = x2, z = y1)  # 3차원 그래프 
# 얘는 scatter를 만듦
surface_2 = go.Scatter3d(x = x1.flatten()[::20],  # x1이 N X N matrix였으니까 이걸 다시 펴줌
                         y = x2.flatten()[::20], # 20개씩 띄어서 
                         z = y2.flatten()[::20], 
                         opacity = 0.3, mode = 'markers')   # opacity는 투명도
data = [surface_1, surface_2]
pyo.iplot(data)
```

```
import numpy as np
import plotly.graph_objs as go  # plotly.graph_objects 라고 full name 안 써도 되나봄
X, Y = np.meshgrid(np.linspace(-4, 4, 50), np.linspace(-3, 3, 50))
X.shape  # 50 x 50 
```

```
Z = np.sin(X**2 + Y**2)  # x^2 + y^2 을 input으로 가지는 sin 그래프
fig = go.Figure(data = [go.Scatter3d(x = X.flatten()[::10], 
                                    y = Y.flatten()[::10], 
                                    z = Z.flatten()[::10], 
                                    opacity = 0.3, mode = "markers")])  
                                    # 어떻게 이렇게 그려지는 지 모르겠음...... 
 fig.show()
```

```
Z = np.sin(X**2, Y**2)
Z1 = Z.copy()
Z1 = Z1 + np.random.rand(50, 50)  # 50 X 50의 random sample 
fig = go.Figure(data = [go.Surface(x = X, y = Y, z = Z), 
                        go.Scatter3d(x = X.flatten()[::10], 
                                     y = Y.flatten()[::10], 
                                     z = Z1.flatten()[::10], 
                                     opacity = 0.3, mode = "markers")])  
                         # 3차원 그래프를 overlap시킬 수 있음.
fig.show()
```

```
Z = np.sin(X**2+Y**2)
Z1 = Z.copy()
Z1 = Z1 + np.random.rand(50, 50)
Z2 = np.sin(X) + np.cos(Y)
fig = go.Figure(data = [go.Surface(x = X, y = Y, z = Z), 
                        go.Scatter3d(x = X.flatten()[::10], 
                                     y = Y.flatten()[::10], 
                                     z = Z1.flatten()[::10], 
                                     opacity = 0.3, mode = "markers"), 
                        go.Surface(x = X, y = Y, z = Z2)])
fig.show()
```
